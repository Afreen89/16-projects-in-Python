```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix
from sklearn import tree
```

```python
df = pd.read_csv("DataSet_Titanic.csv")
df.head()
```

```python
X = df.drop("Survivor", axis=1)
y = df.Survivor
```

```python
X.head()
```

```python
y.head()
```

```python
proj_tree = DecisionTreeClassifier(max_depth=2, random_state=42)
```

```python
proj_tree.fit(X, y)
```

```python
y_pred = proj_tree.predict(X)
print("Accuracy: ", accuracy_score(y_pred, y))
```

```python
confusion_matrix(y, y_pred)
```

```python
plot_confusion_matrix(proj_tree, X, y, cmap=plt.cm.Blues, values_format='.0f')
```

```python
plot_confusion_matrix(proj_tree, X, y, cmap=plt.cm.Blues, values_format='.2f', normalize= "true")
```

```python
plt.figure(figsize= (10, 8))
tree.plot_tree(proj_tree, filled= True, feature_names=X.columns)
plt.show()
```

```python
relevance = proj_tree.feature_importances_
columns = X.columns

sns.barplot(columns, relevance)
plt.title("Relevance")
plt.show()
```
