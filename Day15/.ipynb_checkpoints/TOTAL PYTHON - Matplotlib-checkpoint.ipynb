```python
import matplotlib.pyplot as plt
%matplotlib inline

# Prepare the data
x = list(range(101))
y = []
for number in x:
    y.append(number ** 2)
    
# Prepare the plot area
fig, ax = plt.subplots()

# Add the data
ax.plot(x, y)
ax.set(title= "Acumualated earnings per day", xlabel= "Days", ylabel= "Earnings")

# Save our plot
fig.savefig("/graph-acumulated-revenue.png")
```

```python
import numpy as np

x1 = np.linspace(0, 100, 20)
y1 = x1 ** 2

# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x1, y1)
```

```python
fig, ax = plt.subplots()

x2 = np.linspace(-10, 10, 100)
y2 = np.sin(x2)

ax.scatter(x2, y2)
```

```python
food = {"Pizza": 15, "Burger": 11, "Sandwich": 17}
```

```python
fig, ax = plt.subplots()
ax.bar(food.keys(), food.values())
ax.set(title= "Food Prices", xlabel="Food", ylabel="Prices")
```

```python
fig, ax = plt.subplots()
ax.barh(list(food.keys()), list(food.values()))
ax.set(title= "Food Prices", ylabel= "Food", xlabel= "Prices")
```

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols= 2, figsize= (10, 5))

ax1.plot(x1, y1)
ax2.scatter(x2, y2)
ax3.bar(food.keys(), food.values())
ax4.hist(np.random.randn(1000))
```
